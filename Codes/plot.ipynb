```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

training_24 = pd.read_csv('training_24_loss_new.csv')
training_27 = pd.read_csv('training_27_loss.csv')
training_32 = pd.read_csv('training_32_loss.csv')

validation_24 = pd.read_csv('validation_24_loss_new.csv')
validation_27 = pd.read_csv('validation_27_loss.csv')
validation_32 = pd.read_csv('validation_32_loss.csv')
```

```python
# Plot the data
plt.figure(figsize=(4,3))
plt.plot(training_24["Steps"], training_24["Loss"], linestyle="-", color="b", label="24 GHz")
plt.plot(training_27["Steps"], training_27["Loss"], linestyle="-", color="r", label="27 GHz")
plt.plot(training_32["Steps"], training_32["Loss"], linestyle="-", color="g", label="32 GHz")
#plt.title("Training Loss Curve")
plt.xlabel("Steps")
plt.ylabel('Mean Squared Error')
plt.yscale("log")
plt.legend()
plt.grid(True)
plt.savefig("loss_curve_training.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
# Plot the data
plt.figure(figsize=(4,3))
plt.plot(validation_24["Steps"], validation_24["Loss"], linestyle="-", color="b", label="24 GHz")
plt.plot(validation_27["Steps"], validation_27["Loss"], linestyle="-", color="r", label="27 GHz")
plt.plot(validation_32["Steps"], validation_32["Loss"], linestyle="-", color="g", label="32 GHz")
#plt.title("Training Loss Curve")
plt.xlabel("Steps")
plt.ylabel('Mean Squared Error')
plt.yscale("log")
plt.legend()
plt.grid(True)
plt.savefig("loss_curve_validation.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
